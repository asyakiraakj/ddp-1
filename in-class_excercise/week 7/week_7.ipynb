### Latihan Text Semua Kata Unik

```python
file_name = input("Nama file: ")
# di lowerin aja semuanya biar ga peduli caps

with open(file_name, "r") as f:
    unique_words = []
    for i in range in (len(f.read().split()) - 1):
        words = f.read().split()
        if words[i] != words[i+1]:
           unique_words.append(words)

print(unique_words)
```

```python
file_name = input("Nama file: ")
# di lowerin aja semuanya biar ga peduli caps

with open(file_name, "r") as f:
    unique_words = []
    for words in f.read().split():
        if words.lower() not in unique_words:
           unique_words.append(words)

print(unique_words)
```
