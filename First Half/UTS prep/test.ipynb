### Indexing and Slicing

```python
lst = ["a", "b", "c", "d", "e", "f", "g"]
    #   0    1    2    3    4    5    6
    #  -7   -6   -5   -4   -3   -2   -1

print(lst[-5:-2]) # c, d, e
lst[-2:-5] # not possible krn harusnya dari kanan ke kiri, 
print(lst[-2:-5:-1]) # ini baru bisa karena dari kiri ke kanan
print(lst[::-1]) # reverse string
print(lst[::4]) # dari awal sampe akhir, lompat 4 huruf 
print(lst[::2]) # dari awal sampe akhir, lompat 2 huruf 
print(lst[0:2])
print(lst[:2])
print(lst[:4]) 
print(lst[2:]) 
print(lst[2::3]) 
print(lst[2::-1]) 
```

```python
words = "dfalks@ui.acid dsajlfa;j@ui.ac.id"
nums = "abc1cd"

for word in words.split():
    if ("@" in word) and ("ui.ac.id" in word):
        print(f"@ in word {True}")

for word in words.split():
    if (word in "@") and (word in "ui.ac.id"):
        print(f"word in @ {True}")

if nums in "1234":
    print(f"nums in 1234: {True}")
else:
    print(f"nums in 1234: {False}")

if "1234" in nums:
    print(f"1234 in nums: {True}")
else:
    print(f"1234 in nums: {False}")
```
