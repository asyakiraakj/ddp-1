```python
# UTS Essay no 5 solution
file_name = input("nama file: ")
word = input("sebuah kata: ")

with open(file_name, "r") as file:
    no_baris = 1
    for line in file:
        count = 0
        for words in line.split():
            if words == word:
                count += 1
            # disini perlu mencetak sesuatu, namun bergantung dengan count
            # yaitu ketika count nya 
        if count: # if count > 0
            print(f"baris {no_baris} ada {count} buah kata {word}")
    no_baris += 1
```

Buatlah program yang dapat menampilkan kalimat tanya dari berkas input.txt dan
menyimpannya di berkas output.txt. Sebagai contoh, jika isi dari input.txt seperti di atas,
program akan mengeluarkan berkas output.txt yang berisi:

Apakah kamu sudah makan?
Apa yang terjadi jika es krim tidak dingin?
Soal UTS ini mudah kan?
Setelah ini kamu mau makan es krim?

Suatu kalimat disebut kalimat tanya jika diakhiri dengan tanda tanya “?”. Berkas input.txt
pasti ada, dan minimal berisi satu baris. Setiap barisnya adalah satu kalimat.

```python
with open("input_uts1_4.txt", "r") as file_in,\
     open("output_uts1_4.txt", "w") as file_out:

    for line in file_in:
        if line[-1] == "?":
            print(line, file = "output_uts1_4.txt")

```

```python
with open("input_uts1_4.txt", "r") as file_in,\
     open("output_uts1_4.txt", "w") as file_out:

    for line in file_in:
        line = line.strip() #kalo ga di strip, karakter \n bakal tetep ada jadi di akhir itu semuanya karakter akhrnya \n
        if line[-1] == "?":
            print(line, file = file_out)
```

Latihan Namespace Function

Buat program yang menerima tiga buah integer
lalu tampilkan nilai yang paling maksimal

a = 5
b = 10
c = 4

max = 10

```python
# Latihan Namespace Function

def max2(x,y):
    if x > y:
        return x
    return y # ga pake else, boleh?
             # boleh, karena sekali return itu penghenti fungsi, 
             # jadi kalo misal udh ke execute, sisanya ga ke execute
    
a = int(input("a = "))
b = int(input("b = "))
c = int(input("c = "))

max_num = max2(a, max2(b,c))
#alternative:
max_num_alt = max2(max2(a, b), c)
print(f"maks = {max_num}")
```

# Latihan
Buatlah program yang membaca sebuah file text
Program kemudian menampilkan berapa banyak 
kata yang mengandung huruf vokal



```python
# Solution

def is_contain_vowel(word):
    for char in word:
        if char in "aiueo":
            return True
    return False


file_name = input("Nama file: ")


with open(file_name, "r") as file:
    count = 0
    for word in file.read().split():
        if is_contain_vowel(word):
            count += 1
print(count)
        
```
