```python
class IntCollection():
    def __init__(self):
        self.data = []
        self.index = 0

    def add(self, value):
        self.data.append(value)

    def __next__(self):
        curr_val = self.data[self.index] #note*
        # kapan di stop? 
        if self.index >= len(self.data):
            raise StopIteration
        self.index += 1
        return curr_val


kumpulan_int = IntCollection()
kumpulan_int.add(3)
kumpulan_int.add(4)
kumpulan_int.add(5)
kumpulan_int.add(6)

for e in kumpulan_int:
    print(e)

# ini masih ada error: index out of range
# *: harusnya ini ditaro di bawah, karena dia udh ke akses padahal harusnya gbs,
# * issue kedua: solusi di bawah ini:
if self.index >= len(self.data):
    self.index = 0
```
