#### **COUNTER**

```python
import tkinter as tk

class Counter(tk.Frame):
    def __init__(self, master=None):
        super().__init__(master)
        self.pack()
        self.create_widgets()

    #EVENT HANDLER
    def btn_count_clicked(self, event): #ini baru buat, blm didaftarin
        current_val = int(self.lbl_count["text"]) #ini masih string, ubah ke int
        self.lbl_count["text"] = str(current_val + 1) #dia hanya menerima string, jd ini jadddin strin dlu

    def btn_reset_clicked(self, event):
        self.lbl_count["text"] = "0"

    
    def create_widgets(self):
        self.lbl_count = tk.Label(self, text="0", font = "Times 30")
        self.btn_count = tk.Button(self, text="Count", width=15)
        self.btn_reset = tk.Button(self, text="Reset", width=15)

        self.btn_count.bind("<Button-1>", self.btn_count_clicked)

        self.lbl_count.pack()
        self.btn_count.pack()
        self.btn_reset.pack()

if __name__ == "__main__":
    myApp = Counter()
    myApp.master.title("Aplikasi Counter")
    myApp.master.geometry("300x100")
    myApp.master.mainloop()
```

```python
import tkinter as tk
import random as r

class LoveMeter(tk.Frame):
    def __init__(self, master=None):
        super().__init__(master)
        self.pack()
        self.create_widgets()

    def btn_score_clicked(self, event):
        n1 = self.nama1_var.get()
        n2 = self.nama2_var.get()

        n1 = set(n1)
        n2 = set(n2)

        # score = len(n1 & n2) / len(n1 | n2) * 100
        score = r.random()
        self.lbl_score["text"] = f"{score:.2f}%"

    def create_widgets(self):
        # tambahin instance variabel utk msg" label
        self.nama1_var = tk.StringVar()
        self.nama2_var = tk.StringVar()
        # cara  buat mrk terkait: textVariable

        #Bikin Upper Frame (Frame 1) dan Lower Frame (Frame 2)
        self.upper_frame = tk.Frame(self)
        self.lower_frame = tk.Frame(self)

        self.nama1 = tk.Label(self.upper_frame, text="Nama 1: ")
        self.nama2 = tk.Label(self.upper_frame, text="Nama 2: ")
        self.input_nama1 = tk.Entry(self.upper_frame, text="", textvariable=self.nama1_var)
        self.input_nama2 = tk.Entry(self.upper_frame, text="", textvariable=self.nama2_var)

        self.nama1.grid(row=0, column=0)
        self.nama2.grid(row=1, column=0)
        self.input_nama1.grid(row=0, column=1)
        self.input_nama2.grid(row=1, column=1)

        self.btn_score = tk.Button(self.lower_frame, text="Score:")
        self.lbl_score = tk.Label(self.lower_frame, text="-- %")

        self.btn_score.bind("<Button-1>", self.btn_score_clicked)
        
        self.btn_score.pack()
        self.lbl_score.pack()

        #Masukin Frame 1 dan 2 ke Frame utama
        self.upper_frame.pack()
        self.lower_frame.pack()

if __name__ == "__main__":
    myApp = LoveMeter()
    myApp.master.title("LoveMeter")
    myApp.master.geometry("300x100")
    myApp.master.mainloop()
```

```python
import tkinter as tk
class MyApp(tk.Frame):
    def __init__(self, master = None):
        super().__init__(master)
        self.pack()
        self.lbl_hello = tk.Label(self, text = "Hello!")
        self["width"] = 300
        self["height"] = 200

        self.bind('<Button-1>', self.show_label)
        self.bind('<B1-Motion>', self.show_label)

    def show_label(self, event):
        self.lbl_hello.place(x = event.x, y = event.y)


myapp = MyApp()
myapp.master.geometry("300x200")
myapp.master.mainloop()
```
