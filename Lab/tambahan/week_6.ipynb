```python
# 16

file_name = input("nama file = ")

with open(file_name, "r") as file_in:
    count = 0
    char_per_word = 0
    for word in file_in.read().split():
        char_per_word += len(word)
        count += 1

average = char_per_word/count
print(f"rata-rata panjang kata = {average}")
```

```python
# 17
def count_kar (str, char):
    sum = 0
    for char in str:
        if char == char:
            sum += 1
    return sum

count_kar("halooohalo","o") # kalo di soal yg diminta mendefinisikan fungsi,
                            # ini ga perlu ditulis,
                            # ini gaakan muncil di layar, ini cuma return value
                            # tergantung kita value nya mau diapain,
                            # mau di assign ke variabel kah atau mau di print ke layar
                            # pake print()



```

```python
# 18
file_name = input("nama file = ")

baris_ke = 0
with open(file_name, "r") as file:
    for line in file:
        baris_ke += 1
        vowels = 0 # reset vowels per line
        for char in line:
            if char in "aiueo":
                vowels += 1
        print(f"Baris {baris_ke} : {vowels} huruf vokal")
        
```

```python
# 19
def cari(str, char):
    index = -1
    for characters in str:
        index += 1
        if characters == char:
            return index
    return -1

cari("abcdefghijklmn", "g")

```
