buatlah program yang menerima sebuah string
kemudian program akan menuliskan 2 buah kemungkinan pesan

1. karakter unik
2. karakter tidak unik

unik = kemunculannya hanya sekali

es:
string = halo
karakter unik
string = hallo
karakter tidak unik

hint: gunakan in

```python
str_input = input("string = ")

same_char = 0
for char in str_input:
    if (char in str_input):
        same_char += 1
        
if same_char > 0:
    print("karakter tidak unik")
else:
    print("karakter unik")
```

```python
str_input = input("string = ")

unik = ""
for char in str_input:
    if char not in unik:
        unik += char

if char == unik:
    print("karakter unik")
else:
    print("karakter tidak unik")
```

```python
str_input = input("string = ")

x = ""

for i in str_input:
    if i in x:
        print("tidak unik")
        break
    else:
        x += i
        print("unik")
    
```

```python
str_input = input("string = ")
count = 0

for a in str_input:
    for b in str_input:
        if a == b:
            count += 1
```

```python
# Cara Pa Alfan

string = input("string = ")
is_unik = True

for i in range(len(string)):
    # ex: string = abcdafg
    if string[i] in string[:i]: # string[i] = "a", string[:i] = ""
        is_unik = False
        break

if is_unik:
    print("karakter unik")
else:
    print("karakter tidak unik")
```

buatlah program yang menerima sebuah kata
dan mencetak apakah kata tersebut abecedarian atau tidak
abecedarian: karakter-karakter dari katanya terurut secara alfabetis
contoh: beknow, agnosy, biopsy, acknow

asumsi kata dalam lowercase

hint: huruf bisa dibandingkan

```python
string = input("kata = ")

is_urut = True
for i in range(len(string) - 1): # pake -1
    if string[i] > string[(i+1)]: 
        is_urut = False
    

if is_urut == True:
    print("urut")
else:
    print("tidak urut")
```
