## **Latihan 2**
Definisikan sebuah class Point yang merepresentasikan sebuah titik di bidang kartesian.

Ada dua buah atribut:
1. x: float, posisi x
2. y: float, posisi y

Ada beberapa method:
1. setter dan getter untuk masing-masing atribut
2. method special `__str__(self)`, silakan definisikan sendiri seperti apa representasi stringnya
3. method `distance(other_point)` : return `float`,, jarak dari point `self` ke `other_point` yang juga merupakan point

````python
import math
class Point:
    def __init__(self, x, y):
        self.__x = x
        self.__y = y
    
    def __str__(self):
        return f"Posisi: ({self.__x}, {self.__y})"

    def get_x_position(self):
        return self.__x
    def get_y_position(self):
        return self.__y
    
    def set_x_position(self, val_x):
        self.__x = val_x
    def set_y_position(self, val_y):
        self.__y = val_y

    def distance(self, other_point):
        d_x = (other_point.get_x_position() - self.__x)
        d_y = (other_point.get_y_position() - self.__y)
        return math.sqrt(d_x**2 + d_y **2)

P1 = Point(3, 1)
P2 = Point(0, 4)

dist = P1.distance(P2)
print(dist)
````

Ada 2 Class

atribut:

- title
2. `BookList`
atribut:
- list of books
method:
- `search_book(author)`
- `add_book(author, title)`

````python
class Book:
    def __init__(self, author, title):
        self.author = author
        self.title = title
    def __str__(self):
        return f"Buku berjudul = {self.title} dengan author {self.author}"
    
class BookList:
    def __init__(self):
        self.books = []

    def add_book(self, title, author):
        self.books.append(Book(author, title))

    def search_book(self, author):
        results = []
        for book in self.books:
            if book.author == author:
                results.append(book)
````
