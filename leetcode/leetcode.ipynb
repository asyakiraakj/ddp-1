```python
# TWO SUMS
class Solution:
    def twoSum(self, nums, target: int) -> list[int]:
        ans = []
        for i in range (len(nums)):
            for j in range (i+1, len(nums)):
                if nums[i] + nums[j] == target:
                    ans.append[i]
                    ans.append[j]
        return []
    
nums = [2,7,11,15]
target = 9
Solution.twoSum(nums, target)
```

```python
# PALINDROME NUMBERS
def isPalindrome(x: int) -> bool:
    s = str(x)
    if len(s) <= 2:
        return (s[0] == s[1])
    if s[0] != s[1]:
        return False
    else:
        return isPalindrome(s[1:-1])

isPalindrome(12321)
```
