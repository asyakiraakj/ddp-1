````python
# No 2
# pake 2 list: nyimpen words sama buat count

with open("input.txt", "r") as file:

    ans = []
    for words in file.read().split():
        if words not in ans:
            ans.append(words)

    file.seek(0)
    
    for elem in ans:
        count = 0
        file.seek(0)
        for word in file.read().split():
            if word == elem:
                count += 1
        print(f"{elem}: {count}x")
            
````

````python
# Minta input dari pengguna
input_str = input("Masukkan bilangan ganjil = ")

# Kita gunakan try-except untuk menangani jika pengguna memasukkan selain angka
try:
  # Ubah input string menjadi angka (integer)
  n = int(input_str)

  # Cek apakah bilangan genap
  if n % 2 == 0:
    print("Bilangan harus ganjil")
  # Cek apakah bilangan negatif (karena diminta positif)
  elif n < 1:
     print("Bilangan harus ganjil positif")
  else:
    # --- Input valid (ganjil dan positif), mulai buat pola ---
    
    # Cari indeks baris tengah
    # Contoh: n=5 -> tengah = 5 // 2 = 2
    # Contoh: n=9 -> tengah = 9 // 2 = 4
    tengah = n // 2
    
    # 1. Cetak bagian ATAS (termasuk tengah)
    # Loop dari i = 0 sampai ke 'tengah'
    # Untuk n=5, loop akan berjalan untuk i = 0, 1, 2
    for i in range(tengah + 1):
      # Jumlah bintang adalah 2*i + 1
      # i=0 -> 1 bintang
      # i=1 -> 3 bintang
      # i=2 -> 5 bintang
      print("*" * ( (i * 2) + 1 ))
      
    # 2. Cetak bagian BAWAH (setelah tengah)
    # Loop dari i = (tengah - 1) turun sampai ke 0
    # Untuk n=5, loop akan berjalan untuk i = 1, 0
    for i in range(tengah - 1, -1, -1):
      # Rumus jumlah bintangnya sama
      # i=1 -> 3 bintang
      # i=0 -> 1 bintang
      print("*" * ( (i * 2) + 1 ))

except ValueError:
  # Jika input tidak bisa diubah ke integer (misal: "abc")
  print("Input tidak valid. Harap masukkan angka.")
````
