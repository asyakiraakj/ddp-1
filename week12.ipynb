```python
class SegitigaSikuSiku:
    def __init__(self, alas, tinggi):
        self.__alas = alas
        self.__tinggi = tinggi

    def get_alas(self):
        return self.__alas
    def get_tinggi(self):
        return self.__tinggi

    def get_luas(self):
        return self.__alas * self.__tinggi / 2

    def selisih_luas(self, other:'SegitigaSikuSiku'):
        return self.get_luas() - other.get_luas()
```

```python
s = SegitigaSikuSiku(3., 4.)

print(s.get_alas())
print(s.get_tinggi())
print(s.get_luas())

s2 = SegitigaSikuSiku(3., 8.)

print(s2.selisih_luas(s))
print(s.selisih_luas(s2))
```

Definisikan kelas LabelledPoint yang merepresentasikan sebuah titik di bidang kartesian, dan mempunyai sebuah label (string).

Instance Variables:
• x : int
• y : int
• label : string

Methods: memiliki semua method yang ada di kelas Point + sebuah method baru yaitu, get_label()

```python
import math
class Point:
    def __init__(self, x, y):
        self.__x = x
        self.__y = y
    
    def __str__(self):
        return f"Posisi: ({self.__x}, {self.__y})"

    def get_x_position(self):
        return self.__x
    def get_y_position(self):
        return self.__y
    
    def set_x_position(self, val_x):
        self.__x = val_x
    def set_y_position(self, val_y):
        self.__y = val_y

    def distance(self, other_point):
        d_x = (other_point.get_x_position() - self.__x)
        d_y = (other_point.get_y_position() - self.__y)
        return math.sqrt(d_x**2 + d_y **2)
    

class LabelledPoint(Point):
    def __init__(self, x, y, label):
        super().__init__(x, y)
        self.__label = label

    def get_label(self):
        return self.__label
```
