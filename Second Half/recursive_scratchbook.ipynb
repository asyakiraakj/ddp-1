### Exercise 4: multi
return the product of the distinct elements in the input list (duplicates ignored).

```python
def multi(list):
  unik = []
  for i in list:
    if i not in unik:
      unik.append(i)
  return multi_rec(unik)
    
def multi_rec(new_list):
    if len(new_list)==1:
        return new_list[0]
    return new_list[0]*multi_rec(new_list[1:])
# lst = list(set(lst)) ##cheat##
```

### Exercise 5: fibonacci

```python
def fibonacci(x):
    if x < 1:
        raise ValueError("x harus positif!")
    elif x == 1 or x == 2:
        return 1
    else:
        return fibonacci(x-2) + fibonacci(x-1)
print(fibonacci(7))
# Contributor: Dery Andreas Tampubolon 
```

```python
# iterative version
def fibonacci(n):
    result = 1
    previous_result = 1
    for i in range(3, n + 1):
        temp = result
        result = result + previous_result
        previous_result = temp
    return result

# Contributor: IAF :kekw:
```

### Exercise 6: reverse string

```python
def reverse(str):
    if len(str) == 1:
        return str
    else:
        return str[-1] + reverse(str[:-1])
```

### Exercise 9: deepcopy

```python
# Exercise 9
def deep_copy(list_copy):
    result = []
    for element in list_copy:
        if isinstance(element, list):
            result.append(deep_copy(element))
        elif isinstance(element, tuple):
            result.append(tuple(deep_copy(element)))
        else:
            result.append(element)
    return result
# Contributor: Callista Locytahayu Maesa Naurah Kristanto 2406435225
```

### Exercise 10: flatten complicated list

```python
# Exercise 10
def flatten(complicated_list): 
    result = []
    for element in complicated_list:
        if isinstance(element, list):
            result.extend(flatten(element))
        else:
            result.append(element)
    return result
# Contributor: Ahmad Fauzan Al Ayubi 2406344284
```

### Exercise 11: search

```python
# Exercise 11
def search(item_list, item, path=""):
    for index, value in enumerate(item_list):
        current_path = f"{path},{index}" if path else str(index)
        if value == item:
            return current_path
        if isinstance(value, list):
            result = search(value, item, current_path)
            if result is not None:
                return result
    return None
# Contributor: Sherin Khaira Alya Mirabel 2406404112
```

### Exercise 12: look for children

```python
# Exercise 12
def look_for_children(tree, person):
    for parent, children in tree:
        if parent == person:
            child_list = []
            for child in children:
                child_list.append(child[0])
            return child_list
        else:
            recursive = look_for_children(children, person)
            if recursive is not None:
                return recursive
    return None
# Contributor: Vazha Khayri - 2406495911
```
