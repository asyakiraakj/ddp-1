### **Latihan 1**
Lengkapi definisi `num_printable_bytes(filename)` yang
mengembalikan **integer** <br>yang merepresentasikan berapa banyak bytes
yang dapat dicetak pada sebuah file biner **filename**. <br>
Definisi **byte** yang dapat dicetak adalah <font color="#A5D6FF">“byte yang berkisar dari \x20
(karakter spasi) hingga \x7e (karakter ~)”.<font>


```python
# Ver 1
def num_printable_bytes(filename):
    with open (filename, "rb") as f:
        bs = f.read()
        return sum([1 for b in bs if 0x20 <= b <= 0x7e])

# Ver 2
def num_printable_bytes(filename):
    count = 0
    with open (filename, "rb") as f:
        isi = f.read()
        for ord in isi:
            if 32 <= ord <= 126: #32 = 0x20, 126 = 0x7e. Tipenya sama" integer
                count += 1

    return count
```

### Sempurkanan Fungsi `whos_older`

```python
# Sempurkanan Fungsi whos_older
# isinstance(___, tuple)
# type(...) -> tuple
# INSTROSPECTION

'''
    if x[1] > y[1]:
        return x[0]
    elif x[1] < y[1]:
        return y[0]
    else:
        return None
'''

def whos_older(x, y): 
    if (type(x) != tuple) or (type(y) != tuple):
        raise TypeError("x atau y salah format.")
    
    # perlu ada tambah, kalau x/y sudah tuple, tapi masih salah format
    name_x, age_x = x
    name_y, age_y = y

    if (age_x < 0) or (age_x < 0):
        raise ValueError("umur negatif")

    if x[1] > y[1]:
        return name_x
    elif x[1] < y[1]:
        return name_y
    else:
        return None
```

```python
# def even_sum(a, b, *c): minimal argumen = 2
def even_sum(*args): #minimal argumen = 0
    sum = 0
    for elem in args:
        if elem %2 == 0:
            sum += 1

    return sum
```
