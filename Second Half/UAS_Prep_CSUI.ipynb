##### 5. output dari list comprehension

```python
import math

def factors(number):
    for divisor in range(1, math.floor(math.sqrt(number)) + 1):
        if number % divisor == 0:
            yield number // divisor
            if divisor != (number // divisor):
                yield divisor

print([x for x in factors(36)])
```

```python
#42
def remove_occurances(lst, num):
    for i in range(len(lst)-1):
        if lst[i] == num:
            lst.pop(i)
    return lst

print(remove_occurances([1, 1, 3, 2, 5, -1], 1))
```

```python
#43
def replace_with_star(inp):
    words = inp.strip().split()
    for i in range(len(words) - 1):
        if len(words[i]) > 5:
            words[i] = "*"*len(words[i])
    return " ".join(words)

replace_with_star("GSGS Google Sana Google Sini")
```

```python
def replace_with_star_v2(inp):
    words = inp.strip().split()
    for word in words:
        if len(word) > 5:
            word = "*"*len(word)
    return words

replace_with_star_v2("GSGS Google Sana Google Sini")
```

```python
# 29 MERGE_ALTERNATING
'''
merge_alternating([10, 20, 30], [1, 2, 3])
[10, 1, 20, 2, 30, 3]

merge_alternating([1, 4, 18, 6, 18, 12], [4, 3, 9, 10, 17, 16, 2, 3, 17, 3])
[1, 4, 4, 3, 18, 9, 6, 10, 18, 17, 12, 16, 2, 3, 17, 3]

merge_alternating([4, 2, 5, 3, 10, 16, 2, 4, 20], [3, 3, 14, 6, 19])
[4, 3, 2, 3, 5, 14, 3, 6, 10, 19, 16, 2, 4, 20]
'''

def merge_alternating(lst1, lst2):
    if not lst1:
        return lst1
    if not lst2:
        return lst2
    else:
        if len(lst1) == len(lst2):
            return [lst1[0], lst2[0]] + merge_alternating(lst1[1:], lst2[1:])
'''               ^^^^^^^^^^^^^^^^^^   ^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^
                   ini jadiin list     ditambah rekursi list" selanjutnya
'''
merge_alternating([10, 20, 30], [1, 2, 3])
```

```python
# 30 Merge Sorted
'''
merge_list([1, 1, 2, 3, 4], [1, 2, 3, 4, 5])
[1, 1, 1, 2, 2, 3, 3, 4, 4, 5]
merge_list([10, 40, 90], [5, 8, 9, 11])
[5, 8, 9, 10, 11, 40, 90]
merge_list([1, 50, 100], [5, 10, 15, 20, 25, 30])
[1, 5, 10, 15, 20, 25, 30, 50, 100]
'''
def urutkan(lst):
    for i in range(len(lst)-1):
        if lst[i] > lst[i+1]:
            lst[i+1], lst[i] = lst[i], lst[i+1]
    return lst

def merge_list(lst1, lst2):
    if not lst1:
        return lst1
    if not lst2:
        return lst2
    else:
        return urutkan([lst1[0], lst2[0]] + merge_list(lst1[1:], lst2[1:]))
        
print(merge_list([1, 1, 2, 3, 4], [1, 2, 3, 4, 5]))
print(merge_list([10, 40, 90], [5, 8, 9, 11]))
print(merge_list([1, 50, 100], [5, 10, 15, 20, 25, 30]))
```
