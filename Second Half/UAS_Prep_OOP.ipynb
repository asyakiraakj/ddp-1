```python
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner       # Public (Everyone can see who owns it)
        self._balance = balance  # Protected (Please don't touch this directly!)

my_account = BankAccount("Mario", 100)
print(my_account._balance)
```

```python
class SecretAgent:
    def __init__(self, name, real_identity):
        self.name = name                # Public (anyone can see the codename)
        self.__real_identity = real_identity  # Private (Double underscore!)

james = SecretAgent("007", "James Bond")

# print(james.__real_identity)
print(james.__dict__)
```
