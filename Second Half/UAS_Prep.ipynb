##### 1. **The Palindrome Checker** (Classic Exam Question)<br>
A palindrome is a word that reads the same backward as forward (e.g., "radar", "level"). <br>
Task: Write a function `is_palindrome(s)` that returns `True` or `False`.<br>
- Input: `"racecar"` → Output: `True`
- Input: `"hello"` → Output: `False`
- Hint: Compare the first and last letters. If they match, recurse on the "middle" of the string.

```python
def isPalindrome(s):
    if len(s) <= 1:
        return True
    if s[0] != s[-1]:
        return False
    else:
        return isPalindrome(s[1:-1])
```

##### **2. Sum of Digits** (Math Logic) <br>
Task: Write a function sum_digits(n) that adds up all the digits of a number. <br> 
- Input: 1234
- Output: 10 (because 1 + 2 + 3 + 4 = 10)
- Constraint: You cannot turn the number into a string. You must use math (% and //).
- Hint: n % 10 gives you the last digit. n // 10 removes the last digit.

```python
def sum_digits(n):
    if n == 0:
        return 0
    
    return (n%10) + sum_digits(n//10)
```

##### **3. Climbing Stairs** (LeetCode Style) <br>
You are climbing a staircase. It takes `n` steps to reach the top. Each time you can either climb **1** or **2** steps. In how many distinct ways can you climb to the top? <br>
- Input: 3
- Output: 3
    - Ways: (1+1+1), (1+2), (2+1)

```python
def climbStairs(n):
    # Base Case: If n is 1, return 1
    if n == 1:
        return 1
    # Base Case: If n is 2, return 2
    if n == 2:
        return 2
    
    # Recursive Step: Return sum of (n-1) and (n-2)
    else:
        return climbStairs(n-1) + climbStairs(n-2)
```

##### 5. **Fibonacci**

```python
def fib(n):
    if n == 1:
        return 1
    if n == 0:
        return 0
    else:
        return fib(n-1) + fib(n-2)
```

```python
# ANSWER
filename = input("Enter file name: ")

try:
    with open(filename, 'r') as file:
        # We don't actually need to pre-calculate unique_items! 
        # We can build the dictionary dynamically.
        ans = {}
        
        for line in file:
            line = line.strip()
            if not line: continue  # Skip empty lines
            
            items_this_line = line.split(",")
            
            # For every item in this transaction...
            for key_item in items_this_line:
                # 1. Ensure the key exists in the dictionary
                if key_item not in ans:
                    ans[key_item] = []
                
                # 2. Add all OTHER items from this line to the key's list
                for other_item in items_this_line:
                    # We add it if it's NOT the key_item 
                    # AND if we haven't recorded it yet (to avoid duplicates)
                    if (other_item != key_item) and (other_item not in ans[key_item]):
                        ans[key_item].append(other_item)

        # Print the final result
        for item, others in ans.items():
            print(f"{item}: {others}")

except FileNotFoundError:
    print("File not found.")
```
